```python
# Imports
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
```

# Webscrape Musical Rankings

```python
# Reference: Jacob Ellena - https://github.com/jellena/casting-neural-nets-for-sea-shanties/blob/master/code/scrapers/contemplator-scraper.ipynb
options = Options()
options.headless = True
```

```python
# Credits (selenium, webdriver): https://stackoverflow.com/questions/8049520/web-scraping-javascript-page-with-python

# Get website content with top 197 musicals
url = 'https://www.ranker.com/crowdranked-list/best-musicals'
driver = webdriver.Chrome(executable_path='/usr/local/bin/chromedriver', options=options)
driver.get(url)

timeout = 10

# Credits (presence_of_element_located & view full page): https://stackoverflow.com/questions/26566799/wait-until-page-is-loaded-with-selenium-webdriver-for-python
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
element_present = EC.presence_of_element_located((By.XPATH, '//*[@id="list"]/div[199]'))
WebDriverWait(driver, timeout).until(element_present)

# Credits (ActionChains): https://intellipaat.com/community/26180/selenium-wait-for-page-to-load-selenium-how-to-wait-until-page-is-completely-loaded
element_click = driver.find_element_by_id('pagination__section').find_element_by_id('pagination')
webdriver.ActionChains(driver).move_to_element(element_click).click(element_click).perform()

element_present2 = EC.presence_of_element_located((By.XPATH, '//*[@id="list"]/div[236]'))
WebDriverWait(driver, timeout).until(element_present2)

# Credits (move_to_element): https://stackoverflow.com/questions/3401343/scroll-element-into-view-with-selenium
element2 = driver.find_element_by_xpath('//*[@id="list"]/div[236]')
webdriver.ActionChains(driver).move_to_element(element2).perform()

element_present3 = EC.presence_of_element_located((By.XPATH, '//*[@id="list"]/div[260]'))
WebDriverWait(driver, timeout).until(element_present3)

# Create a dictionary of musicals by rank
content = driver.find_element_by_tag_name('body').find_element_by_tag_name('article').find_elements_by_css_selector('div.listItem')
musical_dict = {}
for musical in content:
    rank = int(musical.find_element_by_tag_name('meta').get_attribute('content'))
    if rank <= 25 and rank not in musical_dict.keys():
        musical_dict[rank] = musical.find_element_by_css_selector('div.listItem__data').find_element_by_tag_name('meta').get_attribute('content')
    elif rank not in musical_dict.keys():
        musical_dict[rank] = musical.find_element_by_css_selector('div.listItem__data').find_elements_by_tag_name('meta')[1].get_attribute('content')

# Exit browser
driver.quit()                                    
```

# Export Musical List

```python
musicals_df = pd.DataFrame(musical_dict.values(), index=musical_dict.keys(), columns=['musical'])
```

```python
musicals_df.to_csv('./data/musical_names.csv')
```
